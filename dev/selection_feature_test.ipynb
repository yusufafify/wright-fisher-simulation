```python
from simulator import WrightFisherSim
```

```python
selection_coeffs = {1: 0.1, 0:0.1}
sim = WrightFisherSim(
    demes_file_path='./deme_test.yml',
    alleles=[0, 1],
    initial_allele_frequency={0: 0.9, 1: 0.1}, # Start with low frequency of beneficial allele
    selection_coefficients=selection_coeffs,
)
```

```python
history = sim.run()

print("\n--- Results: Frequency of Beneficial Allele (1) over time ---")
print(history)
```

```python
# --- 4. Verify Results ---
print("\n--- Results: Frequency of Beneficial Allele (1) over time ---")
pop_history = history['Pop_A']

# The history is stored from start_generation down to 0. 
# We print every 10 generations.
for gen_idx, freqs in enumerate(pop_history):
    # Calculate actual generation number (approximate based on index)
    current_gen = 100 - gen_idx 
    
    if current_gen % 10 == 0:
        freq_1 = freqs.get(1, 0.0)
        print(f"Gen {current_gen}: {freq_1:.4f}")
```
